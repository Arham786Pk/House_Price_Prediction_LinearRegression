```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,3,2,4])
y = np.array([100,300,200,400])
m = x.shape[0]
print(f"X Training Example = {x}")
print(f"Y Training Example = {y}")
print(f"No Of Training Examples = {m}")

i = 2
print(f"x({i}),y({i}) = ( {x[i]} , {y[i]} )")
```

```python
plt.scatter(x,y,marker="x",c="r")
plt.xlabel("Size In Square Feet")
plt.ylabel("Price In $")
plt.title("House Price Prediction")
plt.show()
```

```python
w = 100
b = 0
print(f"w = {w} , b = {b}")
```

```python
def predict_output(x,w,b):
    f = np.zeros(m)
    for i in range(m):
        f[i] = w*x[i] + b
    return f
```

```python
prediction = predict_output(x,w,b)

plt.scatter(x, y, marker="x", c="red", label="Actual Price")
plt.plot(x, prediction , c="blue", label="Our Prediction")

plt.ylabel("Price in $")
plt.xlabel("Size in square feets")
plt.title("House Price Prediction")
plt.legend()
plt.show()
```
